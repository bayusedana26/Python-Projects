```python
# Make network with keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8, input_dim = 4, activation= 'relu')) #8 jumlah neuron, input dim 4 karena ada 4 fitur
model.add(Dense(8, input_dim = 4, activation= 'relu')) #Double layer harapan lebih banyak untuk belajar data

model.add(Dense(3, activation= 'softmax')) #layer untuk output, ada 3 neuron karena ada 3 target

model.compile(loss = 'categorial_crossenthropy', optimizer= 'adam', metrics= ['accuracy'])
```
