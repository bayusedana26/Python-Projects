```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Student': ['Andi', 'Budi', 'Citra', 'Dewi', 'Eko'],
    'Math_Score': [85, 90, np.nan, 88, 76],
    'English_Score': [78, np.nan, 83, 79, np.nan],
    'Science_Score': [np.nan, 91, 89, np.nan, 73]
}

df = pd.DataFrame(data)
print(df)
```

```python
missing_values = df.isnull()
print(missing_values)
```

```python
missing_counts = df.isnull().sum()
print(missing_counts)
```

```python
df_dropped_rows = df.dropna()
print(df_dropped_rows)
```

```python
df_dropped_columns = df.dropna(axis=1)
print(df_dropped_columns)
```
