```python
import re

words = 'MySkill: portal for any learning path'

match = re.search(r'portal', words)

print('Start Index:', match.start())
print('End Index:', match.end())
```

```python
find_dot = "This is a dot ...."

match1 = re.search(r'.', find_dot)

match2 = re.search(r'\.', find_dot)

print(match1)
print(match2)
```

```python
story = "On 4th November I will go to ancol, the ticker price only 25000 Rupiah. It's a discount price. Last time I went to Ancol was in 2 January 2022 and the ticket price was 30000 Rupiah"

number1 = re.compile('\d')
number2 = re.compile('\d+')

print(re.findall(number1, story))

print(re.findall(number2, story))


```
