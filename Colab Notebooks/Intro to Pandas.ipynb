```python
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Saskia'],
    'Math Score': [85, 78, 92, 88, 90],
    'English Score': [95, 85, 89, 76, 88],
    'Science Score': [78, 83, 85, 90, 92]
}

df = pd.DataFrame(data)
df.head()
```

```python
df['Average Score'] = df[['Math Score', 'English Score', 'Science Score']].mean(axis=1)
print(df)
```

```python
high_math_scorers = df[df['Math Score'] > 85]
print(high_math_scorers)
```

```python
sorted_df = df.sort_values(by='Average Score', ascending=False)
print(sorted_df)
```

```python
grouped_df = df.groupby('English Score')['Math Score'].mean()
grouped_df.head()
```
