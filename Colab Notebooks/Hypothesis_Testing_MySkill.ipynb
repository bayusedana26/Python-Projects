## **One Population Hypothesis**

### Manual count

```python
number = [74, 49, 68, 86, 80, 71, 77, 81, 72, 77]
```

```python
x_bar = np.mean(number)
mu = 80
s = np.std(number, ddof =1)
n = len(number)

t_stat = (x_bar - mu) / (s / np.sqrt(n))

print(t_stat)
```

```python
from scipy.stats import t
t.ppf(0.05, n - 1)
```

### Count using python

```python
from scipy.stats import ttest_1samp
```

```python
t_stat, p_value = ttest_1samp(number, 80)
t_stat
```

## **Two Way Hypothesis**

```python
kelas_a = [74, 49, 68, 86, 80, 71, 77, 81, 72, 77]
kelas_b = [78, 57, 65, 82, 73, 76, 54, 58]
```

```python
x_bar_kelasA = np.mean(kelas_a)
s_kelasA = np.std(kelas_a, ddof = 1)
n_kelasA = len(kelas_a)

x_bar_kelasB = np.mean(kelas_b)
s_kelasB = np.std(kelas_b, ddof = 1)
n_kelasB = len(kelas_b)
```

```python
print(f"Standar deviasi untuk kelas a adalah: {s_kelasA}")
print(f"Standar deviasi untuk kelas b adalah: {s_kelasB}")
```

```python
df = n_kelasA + n_kelasB - 2
sp = np.sqrt((( n - 1) * np.power(s_kelasA, 2) + (n_kelasB - 1) * np.power(s_kelasB, 2)) / df)
```

```python
t_stat = (x_bar_kelasA - x_bar_kelasB) / (sp * np.sqrt(1/n_kelasA + 1/n_kelasB))
t_stat
```

### Using Python

```python
from scipy.stats import ttest_ind
```

```python
t_stat, p_value = ttest_ind(kelas_a, kelas_b, equal_var = True)
```

```python
if p_value < 0.05:
  print("Nilai kelas A dan nilai kelas B berbeda")
else:
  print("Nilai kelas A dan nilai kelas B sama")
```

## **T test two paired population**

```python
sebelum = [59, 71, 64, 75, 75, 61, 52, 66, 71, 69]
sesudah = [98, 91, 86, 72, 77, 54, 69, 89, 66, 87]
```

```python
beda = np.array(sesudah) - np.array(sebelum)

x_beda = np.mean(beda)
sd = np.std(beda, ddof = 1)
n_beda = len(beda)
mu = 0
```

```python
t_stat = (x_beda - mu) / (sd / np.sqrt(n_beda))

print(t_stat)
```

### Using Python

```python
from scipy.stats import ttest_rel
```

```python
t_stat, p_value = ttest_rel(sebelum, sesudah)
```

```python
if p_value < 0.05:
  print("Nilai sesudah berbeda daripada sebelum")
else:
  print("Nilai sesudah sama dengan nilai sebelum")
```

## **Test diff proportion**

```python
perempuan_selamat = 233
total_perempuan = 314
laki_selamat = 109
total_laki = 577
```

```python
proporsi_perempuan = perempuan_selamat / total_perempuan
proporsi_laki = laki_selamat / total_laki

prop_gabungan = (perempuan_selamat + laki_selamat) / (total_perempuan + total_laki)
```

```python
t_stat = (proporsi_perempuan - proporsi_laki) / np.sqrt(prop_gabungan * (1 - prop_gabungan) * (1 / total_perempuan + 1 / total_laki))
print(t_stat)
```

### Using Python

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
proportions_ztest([perempuan_selamat, laki_selamat], [total_laki, total_perempuan])
```
