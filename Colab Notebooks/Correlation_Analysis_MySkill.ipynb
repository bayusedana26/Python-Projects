# **Library**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load data

```python
dataset = sns.load_dataset("mpg")
dataset.head()
```

## **Manual**

```python
sns.scatterplot(data = dataset,
                x = "mpg",
                y = "displacement")
```

```python
sns.scatterplot(data = dataset,
                x = "mpg",
                y = "weight")
```

```python
sns.scatterplot(data = dataset,
                x = "acceleration",
                y = "mpg")
```

```python
sns.scatterplot(data = dataset,
                x = "displacement",
                y = "horsepower")
```

## ***Using Python***

```python
correlation = dataset.corr()
correlation
```

```python
sns.heatmap(correlation, vmax = 1, vmin = -1, center = 0)
plt.show()
```

Hypothesis test

```python
from scipy.stats import pearsonr
```

```python
coef, p_val = pearsonr(dataset["mpg"], dataset["displacement"])
```
